```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSV
data = pd.read_csv('Data_Cuaca_Sulteng_Prepared.csv', sep=';')
```

```python
# Visualisasi distribusi temperatur minimum
plt.figure(figsize=(10, 6))
sns.histplot(data['Temperatur minimum (C)'], bins=20, kde=True)
plt.title('Distribusi Temperatur Minimum')
plt.xlabel('Temperatur Minimum (C)')
plt.ylabel('Frekuensi')
plt.grid()
plt.show()
```

```python
# Visualisasi temperatur maksimum
plt.figure(figsize=(10, 6))
sns.histplot(data['Temperatur maximum(C)'], bins=20, kde=True)
plt.title('Distribusi Temperatur Maksimum')
plt.xlabel('Temperatur Maksimum (C)')
plt.ylabel('Frekuensi')
plt.grid()
plt.show()
```

```python
# Visualisasi curah hujan terhadap temperatur rata-rata
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='temperatur rata-rata(C)', y='curah hujan(mm)', alpha=0.6)
plt.title('Curah Hujan vs Temperatur Rata-rata')
plt.xlabel('Temperatur Rata-rata (C)')
plt.ylabel('Curah Hujan (mm)')
plt.grid()
plt.show()
```

```python
# Boxplot kelembapan rata-rata
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='kelembapan rata-rata(%)')
plt.title('Boxplot Kelembapan Rata-rata')
plt.ylabel('Kelembapan Rata-rata (%)')
plt.grid()
plt.show()
```
